# **Ukuran game dalam koleksi platform a dan b**

**Ho**: Perbandingan ukuran game di platform a dan b cenderung lebih sama.

**Ha**: Perbandingan ukuran game di platform a dan b, cenderung lebih banyak di salah satu platform

```python
import pandas as pd
import numpy as np
from scipy import stats
```

# Contoh T Test Single

```python
# Given student scores
student_scores = np.array([72, 89, 65, 73, 79, 94, 93, 96, 55, 95, 10, 90, 70, 78, 93, 60, 97, 88, 90, 98, 95])
df = pd.DataFrame(student_scores)
```

```python
mean = df.mean()
mean = round(mean[0])
```

```python
df.describe()
```

```python
population_scores = 78
```

```python
# Perform one-sample t-test
ttest_result = stats.ttest_1samp(student_scores, population_scores)
ttest_result
# print("T statistic:", ttest_result.statistic)
# print("P-value:", ttest_result.pvalue)
# # Setting significance level
alpha = 0.05
# # Interpret the results
if ttest_result.pvalue <= alpha:
    print("Reject the null Ho; there is a significant difference between the sample mean and the hypothesized population mean.")
else:
    print("Fail to reject the Ho; there is no significant difference between the sample mean and the hypothesized population mean.")
```

# T Test Double Population

```python
platformA = np.array([11, 23, 30, 120, 10, 1, 34, 56, 55, 95, 21, 90, 78, 11, 93, 15, 97, 23, 90, 98, 95])
platformB = np.array([39, 23, 3, 43, 40, 54, 93, 66, 2, 85, 5, 50, 70, 28, 23, 60, 37, 3, 80, 48, 50])
df1 = pd.DataFrame(platformA)
df2 = pd.DataFrame(platformB)
```

```python
mean1 = df1.mean()
mean1 = round(mean1[0])
mean2 = df2.mean()
mean2 = round(mean2[0])

df1.describe()
```

```python
df2.describe()
```

```python
result_ttest_double = stats.ttest_ind(platformA, platformB)
result_ttest_double
print("T statistic:", result_ttest_double.statistic)
print("P-value:", result_ttest_double.pvalue)
```

```python
if result_ttest_double.pvalue <= alpha:
    print("Reject the null Ho; there is a significant difference between the sample mean and the hypothesized population mean.")
else:
    print("Fail to reject the Ho; there is no significant difference between the sample mean and the hypothesized population mean.")
```
