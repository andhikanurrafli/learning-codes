**Pandas Operations**

Creating the DataFrame

```python
df1 = pd.DataFrame({
    'col1':[1,2,3,4],
    'col2':[444,555,666,777],
    'col3':['abc','def','ghi','xyz']
})
df1.head()
```

Apply

```python
def times2(x):
    return x*2
df1['newcol']=df1['col1'].apply(times2)
print(df1)
```

```python
df1['col3Len']=df1['col3'].apply(len)
print(df1)
```

Menambahkan kolom baru (quantity*unitprice)

```python

# Convert 'UnitPrice' column to string type first to handle commas
dfNew['UnitPrice'] = dfNew['UnitPrice'].astype(str)

# Replace commas with periods and convert to float
dfNew['UnitPrice'] = dfNew['UnitPrice'].str.replace(',', '.').astype(float)

# Now you can perform the multiplication
dfNew['TotalAmount'] = dfNew['Quantity'] * dfNew['UnitPrice']
```

```python
dfNew['TotalPrice']=dfNew['Quantity']*dfNew['UnitPrice']
dfNew
```
